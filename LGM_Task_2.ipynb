## **Image to Pencil Sketch**

---



Importing Libraries

```python
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
```

Original Picture

```python
img = cv2.imread('/content/panda.jpg', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

```python
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Converting into black and white image

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

Inverting image pixels

```python
img_invert = cv2.bitwise_not(img_gray)

cv2_imshow(img_invert)
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)

cv2_imshow(img_smoothing)
```

Final pencil sketch

```python
def dodgeV2(x, y):
   return cv2.divide(x, 255 - y, scale=256)
```

```python
final_img = dodgeV2(img_gray, img_smoothing)

cv2_imshow(final_img)
```

Image to pencil sketch for another example

```python
img2 = cv2.imread('/content/unicorn.jpg', cv2.IMREAD_UNCHANGED)
cv2_imshow(img2)
img_gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray2)
img_invert2 = cv2.bitwise_not(img_gray2)
cv2_imshow(img_invert2)
img_smoothing2 = cv2.GaussianBlur(img_invert2, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothing2)
final_img2 = dodgeV2(img_gray2, img_smoothing2)
cv2_imshow(final_img2)
```
